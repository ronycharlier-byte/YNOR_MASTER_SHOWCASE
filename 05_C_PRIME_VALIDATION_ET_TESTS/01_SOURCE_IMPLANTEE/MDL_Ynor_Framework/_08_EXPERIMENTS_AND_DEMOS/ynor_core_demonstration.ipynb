Ce notebook démontre la stabilité thermodynamique d'un système MDL Ynor via les simulations exactes de l'équation $\dot{S} = E(S) - D(S)$.

#### Objectif : Veracité numérique de la Marge Dissipative $\mu$.

```python
import numpy as np
import matplotlib.pyplot as plt
from _04_DEPLOYMENT_AND_API.ynor_core.engine import YnorSystem

# ⚡ FIXATION SEED RIGOUREUSE
np.random.seed(42)

print("✅ Système initialisé - Seed 42 fixée.")
```

### 1. Définition du Système
Nous simulons un système en dimension 2 (ex: Énergie vs Information).

```python
dim = 2
alpha = 2.0  # Force de Dissipation
beta = 0.5   # Force d'Amplification

E = lambda S: beta * S
D = lambda S: alpha * S

sys = YnorSystem(dim, E, D)
S0 = np.array([10.0, 10.0])

mu = sys.measure_dissipative_margin(S0)
print(f"🔥 MARGE DISSIPATIVE CALCULÉE : mu = {mu}")
print(f"📌 Théoriquement mu = alpha - beta = {alpha - beta}")
```

### 2. Simulation Temporelle (Convergence)

```python
t_span = (0, 10)
t_eval = np.linspace(0, 10, 200)

res = sys.simulate(S0, t_span, t_eval)

plt.figure(figsize=(10, 5))
plt.plot(res.t, res.y[0], label="Canal 1 (S1)", color='blue')
plt.plot(res.t, res.y[1], label="Canal 2 (S2)", color='cyan')
plt.title(f"Convergence Exponentielle MDL Ynor (mu={mu})")
plt.xlabel("Temps (t)")
plt.ylabel("Amplitude (S)")
plt.yscale("log")
plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.show()
```
